```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('2.jpg',0) # mở file ảnh
img_equalized = cv2.equalizeHist(img) # cân bằng hist cho ảnh img

fig = plt.figure(figsize=(16, 9))#Tạo vùng vẽ tỷ lệ 16:9
(ax1, ax2), (ax3,ax4) = fig.subplots(2, 2)#Tạo 4 vùng vẽ con, 2 cột 2 hàng

# Vẽ ảnh gốc trong vùng ax1
ax1.imshow(img, cmap='gray')
ax1.set_title("ảnh gốc")

# Vẽ hist của ảnh gốc trong vùng ax2
ax2.hist(img)
ax2.set_title("Histogram ảnh gốc")

# Vẽ ảnh sau khi cân bằng Hist trong vùng ax3
ax3.imshow(img_equalized, cmap='gray')
ax3.set_title("ảnh cân bằng histogram")

# Vẽ hist của ảnh cân bằng hist trong vùng ax4
ax4.hist(img_equalized)
ax4.set_title("cân bằng ảnh ")

plt.show()
```

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('c.jpg', 0)
img_equal_hist = cv2.equalizeHist(img)

f, axes = plt.subplots(2,2, figsize=(30,20))
axes[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title('origin')
axes[0, 1].imshow(cv2.cvtColor(img_equal_hist, cv2.COLOR_BGR2RGB))
axes[0, 1].set_title('hist equal')
axes[1, 0].hist(img.flatten(), 152, [0,152])
axes[1, 1].hist(img_equal_hist.flatten(), 152, [0,125])
plt.show()
plt.close()
```

```python
from google.colab.patches import cv2_imshow
import cv2
img_org = cv2.imread ('2.jpg')
def rescale(frame,scale = 0.5):
    width= int(frame.shape [1]* scale )
    height = int(frame.shape[0]* scale )
    return cv2.resize (frame,(width,height),interpolation=cv2.INTER_AREA)
img = rescale(frame = img_org, scale = 0.2)
#cv2.imshow ('ORG',img)
# anh xam
img_gray = cv2.cvtColor (img,cv2.COLOR_BGR2GRAY )
cv2_imshow ( img_gray )
hist = cv2.calcHist([img_gray],[0],None,[128],[0,128])
plt.figure()
plt.title('histogram')
plt.plot(hist)
plt.show()
```
