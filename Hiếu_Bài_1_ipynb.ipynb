```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('2.jpg',0) # m file nh
img_equalized = cv2.equalizeHist(img) # cn bng hist cho nh img

fig = plt.figure(figsize=(16, 9))#To vng v t l 16:9
(ax1, ax2), (ax3,ax4) = fig.subplots(2, 2)#To 4 vng v con, 2 ct 2 hng

# V nh gc trong vng ax1
ax1.imshow(img, cmap='gray')
ax1.set_title("nh gc")

# V hist ca nh gc trong vng ax2
ax2.hist(img)
ax2.set_title("Histogram nh gc")

# V nh sau khi cn bng Hist trong vng ax3
ax3.imshow(img_equalized, cmap='gray')
ax3.set_title("nh cn bng histogram")

# V hist ca nh cn bng hist trong vng ax4
ax4.hist(img_equalized)
ax4.set_title("cn bng nh ")

plt.show()
cv2_imshow(img)
```
